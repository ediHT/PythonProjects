# Dengue Fever Prediction - Custom Ensemble model
Gradient Boosting Regression/ Gradient Boosting Classification/ GLM Negative Binomial Regression

### Problem Description

Dengue is a mosquito-transmitted viral disease affecting millions of people every year. Accurate dengue predictions would help public health workers and people around the world take steps to reduce the impact of these epidemics.  

So let's get down to it! We have data about Dengue disease cases observed in San Juan (Puerto Rico) and Iquitos (Peru). Also some environment and weather measurements covering:  
- Temperature (minimum, maximum, average and diurnal range)
- Total Precipitation
- Average Dew Point Temperature 
- Average Humidity (relative and specific)
- NDVI  

Normalized difference vegetation index (NDVI):  
- values approaching (-1) - correspond to water
- value approaching 1 - green areas (eg. tropical rainforests)
- values around 0 - rock, sand or snow

#### Approach

This notebook gives a good preview to how one can implement custom ensemble model using GLM Negative Binomial Regression, Gradient Boosting Regression and Gradient Boosted Classification together.

### Feature aggregation
We will create averages of NDVI, precipitation, temperature etc. and also aggregate (average) data by YearMonth and city, since it showed higher stability in the selected modelling approach.

```python
# Let's create average of NDVI, precipitation, temperature etc.
dfAVG = pd.DataFrame()
dfAVG['city'] = df['city']
dfAVG['year'] = df['year']
dfAVG['weekofyear'] = df['weekofyear']
dfAVG['yearMonth']=df['yearMonth']
dfAVG['week_start_date'] = df['week_start_date']
# Feature aggregation/reduction
dfAVG['ndvi_AVG']      = np.nanmean(df[['ndvi_se','ndvi_sw','ndvi_ne','ndvi_nw']], axis=1)
dfAVG['temp_max_AVG']  = np.nanmean(df[['station_max_temp_c', 'reanalysis_max_air_temp_c']], axis=1)
dfAVG['temp_min_AVG']  = np.nanmean(df[['station_min_temp_c', 'reanalysis_min_air_temp_c']], axis=1)
dfAVG['temp_AVG']      = np.nanmean(df[['station_avg_temp_c', 'reanalysis_avg_temp_c','reanalysis_air_temp_c']], axis=1)
dfAVG['temp_diur_AVG'] = np.nanmean(df[['station_diur_temp_rng_c', 'reanalysis_tdtr_k']], axis=1)
dfAVG['precip_AVG']    = np.nanmean(df[['station_precip_mm'
                                        ,'precipitation_amt_mm'
                                        ,'reanalysis_sat_precip_amt_mm']], axis=1)
dfAVG['humidity']      = df['reanalysis_relative_humidity_percent']
dfAVG['dew_point_temp']= df['reanalysis_dew_point_temp_c']
dfAVG['total_cases']   = df['total_cases']
```

```python
# Feature Aggregation by yearMonth and city
for i in range(5, 13, 1):
    dfAVG['ymcAVG_'+ dfAVG.columns[i]] = dfAVG.groupby(['city','yearMonth'])[dfAVG.columns[i]].transform(np.nanmean)
# Additional field (Month)
dfAVG['Month'] = df['Month']

dfAVG.shape
```

## Building the Custom Ensemble Model

There will be 3 parts to this section
1. Gradient Boosting Classifier - to identify the highest outbreaks of Dengue fever
        Classifying training sample observations to 1/0 (I_GBC), where
        1 = top 15% observations having highest Dengue cases
        0 = other
2. Gradient Boosting Regression - adds value to the model, since it predicts the high dengue cases better
3. GLM Negative Binomial Regression - the main model

Ensemble the 3 models
\begin{align}
{y} & = I_{GBC}  (y_{GBR} + y_{NB}) \\
\end{align}

Which means that all the records classified as most likely to have high Dengue cases will get the prediction values from Gradient Boosting Regression, other will be predicted by Negative Binomial Regression.


### Training/ Testing split
- We will have a random train/test sample split (70/30): for model fitting and evaluation.  
    Even though the data is time series (and we are splitting the observations randomly, not by timestamp), let's presume that the relation between environmental measures and dengue cases covers the seasonality.  
- The Predictor Features will be standardised to have mean=0 and variance=1, since a lot of models perform better with standardised values.

## 1. Predicting Dengue cases in San Juan

```python
def TestTrain(Param_city = 'sj'):
    from sklearn.cross_validation import train_test_split
    #Using City indices
    indices = dfKF[dfKF['city']== Param_city].index
    idx_train, idx_test = train_test_split(indices, test_size = 0.3, random_state = 0)
    return idx_train, idx_test
#------------
idx_train, idx_test = TestTrain(Param_city = 'sj')
```

```python
idx_train.shape
```

### 1.1. Extreme Dengue cases classification model (Top 15% highest Dengue cases)
Using Gradient Boosted Classifier

```python
def GBCmod(idx_train, idx_test):
    Yq85 = dfKF.iloc[idx_train, 13].quantile(0.85)
    y0 = np.where(dfKF['total_cases'] < Yq85, 0, 1)


    # Will use the same train/test split for all models so we can ensemble them at the end

    X_train0 = dfKF.iloc[idx_train, start_col_train:].values
    y_train0 = y0[idx_train]
    X_test0 = dfKF.iloc[idx_test, start_col_train:].values
    y_test0 = y0[idx_test]

    # Standardize data using standardScaler (mean=0, std=1 ?)
    sc_X = StandardScaler()
    X_train0 = sc_X.fit_transform(X_train0)
    X_test0 = sc_X.transform(X_test0)

    #-----------------------------------------------------
    # Gradient Boosted Classifier
    from sklearn.ensemble import GradientBoostingClassifier
    GBC = GradientBoostingClassifier(learning_rate=0.1, loss='deviance', max_depth=7,
                                      min_samples_leaf=7, min_samples_split=14,
                                      n_estimators=100, random_state=None, verbose=0)

    GBC.fit(X_train0, y_train0)

    # Make the prediction on the Testing sample
    y_pred0 = GBC.predict(X_test0)
    y_pred_train0 = GBC.predict(X_train0)
    y_pred_prob1 = GBC.predict_proba(X_test0)[:,1]

    #------------------------------------------
    # MODEL EVALUATION
    #------------------------------------------
    from sklearn import metrics
    #---------------------
    # Confusion matrix
    # Table that describes the performance of a classification model
    confusion = metrics.confusion_matrix(y_test0, y_pred0)
    print('          Predicted = 0    Predicted = 1')
    print('Actual(0)      ' + str(confusion[0, 0]) + '              ' + str(confusion[0, 1]))
    print('Actual(1)      ' + str(confusion[1, 0]) + '              ' + str(confusion[1, 1]))
    print('\n')
    #[row, column]
    TP = confusion[1, 1]
    TN = confusion[0, 0]
    FP = confusion[0, 1]
    FN = confusion[1, 0]

    # Classification Error: Overall, how often is the classifier incorrect? ((FP + FN) /All)
    # AUC tells how much model is capable of distinguishing between classes
    print('Proportion of Correct Predictions (Prediction Accuracy) and coefficient of determination:')
    print('On Training sample = %f ' % (metrics.accuracy_score(y_train0, y_pred_train0)))
    print('r-squared          = %f' % (r2_score(y_train0, y_pred_train0)))
    print('\n')
    print('On Testing sample  = %f' % (metrics.accuracy_score(y_test0, y_pred0)))
    print('r-squared          = %f' % (r2_score(y_test0, y_pred0)))
    print('\n')    
    print('Classification Error: ' + str(1 - metrics.accuracy_score(y_test0, y_pred0)))
    print('Rate of Pred(1) vs. Actual(1) : ' + str(TP / (TP + FN)))
    print('AUC (Percentage of ROC plot underneath the curve): ' + str(metrics.roc_auc_score(y_test0, y_pred_prob1)))

    #--------------------------------------  
    # Plot feature importance
    feature_importance = GBC.feature_importances_
    # make importances relative to max importance
    feature_importance = 100.0 * (feature_importance / feature_importance.max())
    sorted_idx = np.argsort(feature_importance)
    # Let's plot to top10 most important variables
    sorted_idx = sorted_idx[-10:]
    pos = np.arange(sorted_idx.shape[0]) + .5
    plt.barh(pos, feature_importance[sorted_idx], align='center')
    plt.yticks(pos, dfKF.columns[sorted_idx + start_col_train])
    plt.xlabel('Relative Importance')
    plt.title('Variable Importance');
    
    #------
    return y_pred0
#**********************
y_pred0_sj = GBCmod(idx_train, idx_test)
```

### 1.2. Gradient Boosted Regression model - Dengue cases prediction (San Juan)

```python
def GBRmod(idx_train, idx_test):

    X_train = dfKF.iloc[idx_train, start_col_train:].values
    y_train = dfKF.iloc[idx_train, 13].values
    X_test = dfKF.iloc[idx_test, start_col_train:].values
    y_test = dfKF.iloc[idx_test, 13].values

    # Standardize data using standardScaler (mean=0, std=1 ?)
    sc_X = StandardScaler()
    X_train = sc_X.fit_transform(X_train)
    X_test = sc_X.transform(X_test)
    
    #------------------------------------------
    # Setting parameters and fitting the model
    #------------------------------------------
    GBR = GradientBoostingRegressor(loss='ls',                      
                                    n_estimators=150, max_depth=5,
                                    learning_rate=.1, min_samples_leaf=14,
                                    min_samples_split=28)
    GBR.fit(X_train, y_train)

    # Make the prediction on the Testing sample (and score training sample to see how well it was fitted)
    y_pred = GBR.predict(X_test)
    y_pred_train = GBR.predict(X_train)
    
    #------------------------------------------
    # MODEL EVALUATION
    #------------------------------------------
    print('Model Accuracy on Training sample:')
    print('Mean Absolute Error (MAE) = %f' % (mean_absolute_error(y_train, y_pred_train)))
    print('r-squared                 = %f' % (r2_score(y_train, y_pred_train)))
    print('\n')
    
    print('Accuracy of prediction on Testing sample:')
    print('MAE         = %f' % (mean_absolute_error(y_test, y_pred)))
    print('r-squared   = %f' % (r2_score(y_test, y_pred))) 
    print('\n') 
    
    DFyy = pd.DataFrame({'week_start_date':dfKF.iloc[idx_test,4],'y_test':y_test,'y_pred': y_pred})
    DFyy.sort_values(by=['week_start_date'],inplace=True)
    q75test = DFyy['y_test'].quantile(0.75)
    print('Accuracy of prediction on Top 25% values of Testing sample:')
    print('MAE         = %f' % (mean_absolute_error(DFyy[DFyy['y_test']> q75test]['y_test'], \
                                                    DFyy[DFyy['y_test']> q75test]['y_pred'])))
    q25test = DFyy['y_test'].quantile(0.25)
    print('Accuracy of prediction on Bottom 25% values of Testing sample:')
    print('MAE         = %f' % (mean_absolute_error(DFyy[DFyy['y_test'] < q25test]['y_test'], \
                                                    DFyy[DFyy['y_test'] < q25test]['y_pred'])))

    #--------------------------------------  
    # Plot feature importance
    feature_importance = GBR.feature_importances_
    # make importances relative to max importance
    feature_importance = 100.0 * (feature_importance / feature_importance.max())
    sorted_idx = np.argsort(feature_importance)
    # Let's plot to top10 most important variables
    sorted_idx = sorted_idx[-10:]
    pos = np.arange(sorted_idx.shape[0]) + .5
    plt.barh(pos, feature_importance[sorted_idx], align='center')
    plt.yticks(pos, dfKF.columns[sorted_idx + start_col_train])
    plt.xlabel('Relative Importance')
    plt.title('Variable Importance');
    
    #--------------------------------------
    # Plot actual vs. predicted
    fig = plt.figure(figsize=(10, 5))
    plt.plot(DFyy['week_start_date'], DFyy['y_test'], label='Actual')
    plt.plot(DFyy['week_start_date'], DFyy['y_pred'], 'm', label='Predicted', alpha=0.5)
    plt.xlabel('Week Start Date')
    plt.title('Predicted vs. Actual Dengue Cases')
    plt.legend(loc='upper right');
    
    #--------------------------------------------
    # Plot Actual vs. Predicted Dengue cases (Values sorted by actual values)
    DFyy.sort_values(by=['y_test'],inplace=True)
    fig = plt.figure(figsize=(10, 5))
    plt.plot(np.arange(0,len(DFyy),1), DFyy['y_test'],  label='Actual')
    plt.plot(np.arange(0,len(DFyy),1), DFyy['y_pred'], 'm', label='Predicted', alpha=0.5)
    #plt.ylim(0,500000)
    plt.xlabel('Observation')
    plt.title('Predicted vs. Actual Dengue Cases (Observations sorted by actual dengue values)')
    plt.legend(loc='upper right');
    
    #------
    return y_pred
#**********************
y_pred_sj = GBRmod(idx_train, idx_test)
```

>  This method is sensitive to predictors value change, which can lead to overfitting. Would be sensible to do cross-validation to check model stability

```python
def get_best_model(idx_train, idx_test):

    df_Xy_train = dfKF.loc[idx_train, ['ymcAVG_ndvi_AVG', 'ymcAVG_temp_min_AVG', 'ymcAVG_humidity', \
                                       'ymcAVG_precip_AVG','ymcAVG_dew_point_temp','total_cases']]
    df_Xy_test = dfKF.loc[idx_test, ['ymcAVG_ndvi_AVG', 'ymcAVG_temp_min_AVG', 'ymcAVG_humidity', \
                                     'ymcAVG_precip_AVG','ymcAVG_dew_point_temp','total_cases']]
    
    # Step 1: specify the form of the model
    model_formula = "total_cases ~ 1 + " \
                    "ymcAVG_ndvi_AVG + " \
                    "ymcAVG_temp_min_AVG + " \
                    "ymcAVG_humidity + " \
                    "ymcAVG_precip_AVG + " \
                    "ymcAVG_dew_point_temp"


    grid = 10 ** np.arange(-8, -1, dtype=np.float64)
    best_alpha = []
    best_score = 1000

    # Step 2: Find the best hyper parameter, alpha
    for alpha in grid:
        model = smf.glm(formula=model_formula,
                        data=df_Xy_train,
                        family=sm.families.NegativeBinomial(alpha=alpha))

        results = model.fit()
        predictions = results.predict(df_Xy_test).astype(int)
        score = eval_measures.meanabs(predictions, df_Xy_test.total_cases)

        if score < best_score:
            best_alpha = alpha
            best_score = score

    print('best alpha = ', best_alpha)
    print('best score = ', best_score)

    # Step 3: final fitting
    model = smf.glm(formula=model_formula,
                    data=df_Xy_train,
                    family=sm.families.NegativeBinomial(alpha=best_alpha))
    fitted_model = model.fit()
    y_pred = fitted_model.predict(df_Xy_test).astype(int)
    y_pred_train =  fitted_model.predict(df_Xy_train).astype(int)
    
    
    #-------------- EVALUATION --------------------------
    print('\nGLM Model Accuracy on Training sample:')
    print('MAE         = %f' % (mean_absolute_error(df_Xy_train['total_cases'], y_pred_train)))
    print('r-squared   = %f' % (r2_score(df_Xy_train['total_cases'], y_pred_train)))

    print('\nAccuracy of prediction on Testing sample:')
    print('MAE         = %f' % (mean_absolute_error(df_Xy_test['total_cases'], y_pred)))
    print('r-squared   = %f' % (r2_score(df_Xy_test['total_cases'], y_pred)))

    
    DFyy = pd.DataFrame({'week_start_date':dfKF.iloc[idx_test,4],'y_test':df_Xy_test['total_cases'],'y_pred': y_pred})
    DFyy.sort_values(by=['week_start_date'],inplace=True)  

    q75test = DFyy['y_test'].quantile(0.75)
    print('Accuracy of prediction on Top 25% values of Testing sample:')
    print('MAE         = %f' % (mean_absolute_error(DFyy[DFyy['y_test']> q75test]['y_test'], \
                                                    DFyy[DFyy['y_test']> q75test]['y_pred'])))
    q25test = DFyy['y_test'].quantile(0.25)
    print('Accuracy of prediction on Bottom 25% values of Testing sample:')
    print('MAE         = %f' % (mean_absolute_error(DFyy[DFyy['y_test'] < q25test]['y_test'], \
                                                    DFyy[DFyy['y_test'] < q25test]['y_pred'])))
    #--------------------------------------
    # Plot actual vs. predicted
    fig = plt.figure(figsize=(10, 5))
    plt.plot(DFyy['week_start_date'], DFyy['y_test'], label='Actual')
    plt.plot(DFyy['week_start_date'], DFyy['y_pred'], 'm', label='Predicted', alpha=0.5)
    plt.xlabel('Week Start Date')
    plt.title('Predicted vs. Actual Dengue Cases')
    plt.legend(loc='upper right');
    #------
    return y_pred
#*******************

y_pred_NB_sj = get_best_model(idx_train, idx_test)
```

#### Concluding Comment 
Clearly the results here suggest further investigation and model improvement.  
The custom ensemble model with GLM Negative Binomial Regression was created due to results on actual testing sample submitted to the drivendata competition site. The Submission of predictions based on Gradient Boosted Regression (which has quite good MAE on our testing sample here) got mean absolute error (MAE) = 26 at submission sample, where the custom ensemble model got MAE = 23.8. So the Gradient Boosting model is most likely overfitting, since the method is tree-based it is sensitive to predictor value changes and fore our case doesn't predict well small dengue cases. The overfitting usually is sorted by hyper-parameter tuning. There's also a chance that original testing/training sample split is related to these issues, since our train/validation split gives similar results of MAE.
